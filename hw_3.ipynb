# Линейные преобразования.

### Задание №1.
Найти собственные векторы и собственные значения для линейного оператора, заданного матрицей

$$A=\begin{pmatrix}
-1 & -6\\ 
2 & 6
\end{pmatrix}.$$

__Решение:__

Составим характеристическое уравение и найдем $\lambda$:

$$det(A-\lambda I)=0$$

$$\begin{vmatrix} -1-\lambda & -6 \\ 2 & 6-\lambda \end{vmatrix} = 0$$

$$(-1-\lambda)(6-\lambda) + 12 = 0$$

$$\lambda^2 - 5\lambda + 6 = 0$$

$$(\lambda - 2.5)^2 - 6.25 + 6 = 0$$

$$(\lambda - 2.5)^2 = 0.25$$

$$\lambda_1 = 2\,\,\,\,\,\lambda_2 = 3$$

Для каждого собственного числа найдем собственный вектор:

1. $\lambda = 2$

$$\begin{pmatrix} -1 & -6 \\ 2 & 6 \end{pmatrix}\cdot\begin{pmatrix} x_1 \\ y_1 \end{pmatrix} = 2\cdot\begin{pmatrix} x_1 \\ y_1 \end{pmatrix}$$

$$\begin{cases} -x_1-6y_1=2x_1 \\ 2x_1+6y_1=2y_1 \end{cases}$$

$$\begin{cases} x_1=-2y_1 \\ x_1=-2y_1 \end{cases}$$

Получаем множество векторов, параллельных прямой $x+2y=0$. В частности вектор $(2, -1)$.

2. $\lambda = 3$

$$\begin{pmatrix} -1 & -6 \\ 2 & 6 \end{pmatrix}\cdot\begin{pmatrix} x_2 \\ y_2 \end{pmatrix} = 3\cdot\begin{pmatrix} x_2 \\ y_2 \end{pmatrix}$$

$$\begin{cases} -x_2-6y_2=3x_1 \\ 2x_2+6y_2=3y_2 \end{cases}$$

$$\begin{cases} x_1=-1.5y_1 \\ x_1=-1.5y_1 \end{cases}$$

Для этого собственного числа получаем множество векторов, параллельных прямой $x+1.5y=0$. В частности вектор $(3, -2)$

__Ответ:__   
$\lambda_1=2,\,\,\,(2, -1)$   
$\lambda_2=3,\,\,\,(3, -2)$

### Задание №2.
Дан оператор поворота на 180 градусов, задаваемый матрицей 

$$A=\begin{pmatrix}
-1 & 0\\ 
0 & -1
\end{pmatrix}.$$

Показать, что __любой__ вектор является для него собственным.

__Решение:__

Поворот вектора на 180 градусов по сути обозначает умножение его на -1. -1 и является собственным числом.   
Составим уравнение для поиска собственного вектора:

$$\begin{pmatrix} -1 & 0 \\ 0 & -1 \end{pmatrix}\cdot\begin{pmatrix} x \\ y \end{pmatrix} = -1\cdot\begin{pmatrix} x \\ y \end{pmatrix}$$

$$\begin{cases} -x=-x \\ -y=-y \end{cases}$$

Под это решение попадает абсолютно все векторы двумерного пространства.

### Задание №3.
Пусть линейный оператор задан матрицей

$$A=\begin{pmatrix}
1 & 1\\ 
-1 & 3
\end{pmatrix}.$$

Установить, является ли вектор $x=(1,1)$ собственным вектором этого линейного оператора.

__Решение:__

Предположим, что вектор $x$ является собственным вектором заданного линейного оператора, тогда должно существовать некоторое вещественное число $\lambda$, при котором

$$\begin{pmatrix} 1 & 1 \\ -1 & 3 \end{pmatrix}\cdot\begin{pmatrix} 1 \\ 1 \end{pmatrix} = \lambda\cdot\begin{pmatrix} 1 \\ 1 \end{pmatrix}$$

$$\begin{cases} 2=\lambda \\ 2=\lambda \end{cases}$$

Решение однозначное, следовательно вектор $x=(1,1)$ __является собственным__ для заданного оператора.

### Задание №4.
Пусть линейный оператор задан матрицей

$$A=\begin{pmatrix} 0 & 3 & 0\\ 3 & 0 & 0\\ 0 & 0 & 3 \end{pmatrix}.$$

Установить, является ли вектор $x=(3, -3, -4)$ собственным вектором этого линейного оператора.

__Решение:__

Аналогично предыдущей задаче предположим, что вектор $x$ является собственным вектором заданного линейного оператора, тогда должно существовать некоторое вещественное число $\lambda$, при котором

$$\begin{pmatrix} 0 & 3 & 0\\ 3 & 0 & 0\\ 0 & 0 & 3 \end{pmatrix}\cdot\begin{pmatrix} 3 \\ -3 \\ -4 \end{pmatrix} = \lambda\cdot\begin{pmatrix} 3 \\ -3 \\ -4 \end{pmatrix}$$

$$\begin{cases} -9=3\lambda \\ 9=-3\lambda \\ -12=-4\lambda \end{cases}$$

$$\begin{cases} \lambda=-3 \\ \lambda=-3 \\ \lambda=3 \end{cases}$$

Возникло противоречие, следовательно вектор $x=(3,-3,-4)$ __не является собственным__ для заданного оператора.
