### Задание №1.
Решить систему уравнений методом Гаусса:

$$\begin{cases}
x_{1}+x_{2}-x_{3}-2x_{4}=0, \\
2x_{1}+x_{2}-x_{3}+x_{4}=-2, \\
x_{1}+x_{2}-3x_{3}+x_{4}=4.
\end{cases}$$

__Решение:__

Составим расширенную матрицу системы:

$$\begin{pmatrix}1&1&-1&-2&0 \\ 2&1&-1&1&-2 \\ 1&1&-3&1&4\end{pmatrix}$$

Приведем ее к ступенчатой, выполнив преобразования. Из второй строки вычтем первую, умноженную на 2, из третьей строки вычтем первую:

$$\begin{pmatrix}1&1&-1&-2&0 \\ 2&1&-1&1&-2 \\ 1&1&-3&1&4\end{pmatrix}\,\,\,\Rightarrow\,\,\,\begin{pmatrix}1&1&-1&-2&0 \\ 0&-1&1&5&-2 \\ 0&0&-2&3&4\end{pmatrix}$$

Получившаяся матрица соответствует системе:

$$\begin{cases}
x_{1}+x_{2}-x_{3}-2x_{4}=0, \\
-x_{2}+x_{3}+5x_{4}=-2, \\
-2x_{3}+3x_{4}=4.
\end{cases}$$

И так как количество уравнений меньше количества неизвестных, то система имеет бесконечное множество решений. Запишем решение в общем виде, приняв $x_4 = c$:

$$\begin{cases} x_4 = c \\ x_3 = 1.5c-2 \\ x_2 = 6.5c \\ x_1 = -3c-2 \end{cases}$$

Подставляя вместо $c$ любое число, получаем частное решение.

### Задание №2.
Проверить на совместность и выяснить, сколько решений будет иметь система линейных уравнений:

   а) $\begin{cases}
3x_{1}-x_{2}+x_{3}=4, \\
2x_{1}-5x_{2}-3x_{3}=-17, \\
x_{1}+x_{2}-x_{3}=0;
\end{cases}$
    
   б) $\begin{cases}
2x_{1}-4x_{2}+6x_{3}=1, \\
x_{1}-2x_{2}+3x_{3}=-2, \\
3x_{1}-6x_{2}+9x_{3}=5;
\end{cases}$
    
   в) $\begin{cases}
x_{1}+2x_{2}+5x_{3}=4, \\
3x_{1}+x_{2}-8x_{3}=-2. 
\end{cases}$

__Решение:__

__а)__ Посчитаем определитель матрицы коэффициентов:

$$\begin{vmatrix}3&-1&1 \\ 2&-5&-3 \\ 1&1&-1\end{vmatrix} = 3\cdot(-5)\cdot(-1) + (-1)\cdot(-3)\cdot1 + 1\cdot2\cdot1 - 1\cdot(-5)\cdot1 - (-1)\cdot2\cdot(-1) - 3\cdot(-3)\cdot1 = 15 + 3 + 2 + 5 - 2 + 9 = 32$$

Значит ранг матрицы коэффициентов, как и ранг расширенной матрицы, равен количеству неизвестных (3). Это говорит нам о том, что система совместна и имеет единственное решение.

__б)__ Векторы коэффициентов $\begin{pmatrix}2\\1\\3\end{pmatrix},\,\,\begin{pmatrix}-4\\-2\\-6\end{pmatrix},\,\,\begin{pmatrix}6\\3\\9\end{pmatrix}$ линейно зависимы и лежат на одной прямой, на которой __не располагается__ вектор свободных членов $\begin{pmatrix}1\\-2\\5\end{pmatrix}$. Следовательно его нельзя разложить по векторам коэффициентов и система уравнений несовместна.

__в)__ Ранг матрицы коэффициентов, как и ранг расширенной матрицы, равен 2, и он меньше количества неизвестных. Следовательно данная система совместна и имеет бесконечное количество решений.

### Задание №1.
Решить систему уравнений методом Крамера:

   а) $\begin{cases}
x_{1}-2x_{2}=1 \\
3x_{1}-4x_{2}=7
\end{cases}$
    
   б) $\begin{cases}
2x_{1}-x_{2}+5x_{3}=10 \\
x_{1}+x_{2}-3x_{3}=-2 \\
2x_{1}+4x_{2}+x_{3}=1
\end{cases}$

__Решение:__

__а)__ Найдем определитель матрицы коэффициентов:

$\det A = \begin{vmatrix} 1&-2 \\ 3&-4 \end{vmatrix} = 1\cdot(-4) - (-2)\cdot3 = 2$.

Система совместна. Найдем определители $\det A_i$:

$\det A_1 = \begin{vmatrix} 1&-2 \\ 7&-4 \end{vmatrix} = 1\cdot(-4) - (-2)\cdot7 = 10$;

$\det A_2 = \begin{vmatrix} 1&1 \\ 3&7 \end{vmatrix} = 1\cdot7 - 1\cdot3 = 4$.

Тогда

$x_1 = \displaystyle\frac{\det A_1}{\det A} = 5$;

$x_2 = \displaystyle\frac{\det A_2}{\det A} = 2$.

__б)__ Найдем определитель матрицы коэффициентов:

$\det A = \begin{vmatrix} 2&-1&5 \\ 1&1&-3 \\ 2&4&1 \end{vmatrix} = 2\cdot1\cdot1 + (-1)\cdot(-3)\cdot2 + 5\cdot1\cdot4 - 5\cdot1\cdot2 - (-1)\cdot1\cdot1 - 2\cdot(-3)\cdot4 = 43$.

Система совместна. Найдем определители $\det A_i$:

$\det A_1 = \begin{vmatrix} 10&-1&5 \\ -2&1&-3 \\ 1&4&1 \end{vmatrix} = 10\cdot1\cdot1 + (-1)\cdot(-3)\cdot1 + 5\cdot(-2)\cdot4 - 5\cdot1\cdot1 - (-1)\cdot(-2)\cdot1 - 10\cdot(-3)\cdot4 = 86$;

$\det A_2 = \begin{vmatrix} 2&10&5 \\ 1&-2&-3 \\ 2&1&1 \end{vmatrix} = 2\cdot(-2)\cdot1 + 10\cdot(-3)\cdot2 + 5\cdot1\cdot1 - 5\cdot(-2)\cdot2 - 10\cdot1\cdot1 - 2\cdot(-3)\cdot1 = -43$;

$\det A_3 = \begin{vmatrix} 2&-1&10 \\ 1&1&-2 \\ 2&4&1 \end{vmatrix} = 2\cdot1\cdot1 + (-1)\cdot(-2)\cdot2 + 10\cdot1\cdot4 - 10\cdot1\cdot2 - (-1)\cdot1\cdot1 - 2\cdot(-2)\cdot4 = 43$.

Тогда

$x_1 = \displaystyle\frac{\det A_1}{\det A} = 2$;

$x_2 = \displaystyle\frac{\det A_2}{\det A} = -1$;

$x_3 = \displaystyle\frac{\det A_3}{\det A} = 1$.

### Задание №3.
Решить систему линейных уравнений методом $LU$-разложения

$$\begin{cases}
2x_{1}+x_{2}+3x_{3}=1 \\
11x_{1}+7x_{2}+5x_{3}=-6 \\
9x_{1}+8x_{2}+4x_{3}=-5
\end{cases}$$

__Решение:__

Представим систему в матричном виде:

$$\begin{pmatrix}
2 & 1 & 3 \\
11 & 7 & 5 \\
9 & 8 & 4
\end{pmatrix}\cdot
\begin{pmatrix}
x_1 \\
x_2 \\
x_3
\end{pmatrix}=
\begin{pmatrix}
1 \\
-6 \\
-5
\end{pmatrix}$$

Матрица может быть разложена $LU$разложением единственным образом, если она не вырождена и её главные миноры не равны $0$. Проверим:

```python
import numpy as np
a = np.array([[2, 1, 3], [11, 7, 5], [9, 8, 4]])
for i in range(2, -1, -1):
    print(round(np.linalg.det(a[:i+1, :i+1]), 3))
```

Разложим матрицу коэффициентов на $L$ и $U$:

- вычтем из второй строки первую, умноженную на $5.5$, из третьей строки вычтем первую, умноженную на $4.5$. Для матрицы $L$ получим $l_{21}=5.5,\,l_{31}=4.5$, а матрица $U$ примет вид:

$$\begin{pmatrix}
2 & 1 & 3 \\
0 & 1.5 & -11.5 \\
0 & 3.5 & -9.5
\end{pmatrix}.$$

- далее из третьей строки вычтем вторую, умноженную на $7/3$. Для матрицы $L$ получим $l_{32}=7/3$, а матрица $U$ примет вид:

$$\begin{pmatrix}
2 & 1 & 3 \\
0 & 1.5 & -11.5 \\
0 & 0 & 52/3
\end{pmatrix}.$$

В итоге:
$$L = \begin{pmatrix}
1 & 0 & 0 \\
5.5 & 1 & 0 \\
4.5 & 7/3 & 1
\end{pmatrix},$$

$$U = \begin{pmatrix}
2 & 1 & 3 \\
0 & 1.5 & -11.5 \\
0 & 0 & 52/3
\end{pmatrix}.$$

Решим теперь систему 

$$Ly=b:$$

$$\begin{cases}
y_{1}=1, \\
5.5y_{1}+y_{2}=-6, \\
4.5y_{1}+(7/3)y_{2}+y_{3}=-5.
\end{cases}$$

$$y_{1}=1,$$
$$y_{2}=-11.5,$$
$$y_{3}=52/3.$$

И затем систему

$$Ux=y:$$

$$\begin{cases}
2x_{1}+x_{2}+3x_{3}=1, \\
1.5x_{2}-11.5x_{3}=-11.5, \\
(52/3)x_{3}=52/3.
\end{cases}$$

$$x_{3}=1,$$
$$x_{2}=0,$$
$$x_{1}=-1.$$

### Задание №4.
Решить систему линейных уравнений методом Холецкого

$$\begin{cases}
81x_{1}-45x_{2}+45x_{3}=531 \\
-45x_{1}+50x_{2}-15x_{3}=-460 \\
45x_{1}-15x_{2}+38x_{3}=193
\end{cases}$$

__Решение:__

$$\begin{pmatrix}
81 & -45 & 45 \\
-45 & 50 & -15 \\
45 & -15 & 38
\end{pmatrix}\cdot
\begin{pmatrix}
x_1 \\
x_2 \\
x_3
\end{pmatrix}=
\begin{pmatrix}
531 \\
-460 \\
193
\end{pmatrix}$$

Матрица может быть разложена по методe Холецкого единственным образом, если она симмметрична, невырождена, её главные миноры не равны $0$ и собственные значения больше $0$. Очевидно первому условию матрица удовлетворяет. Проверим оставшиеся:

```python
a = np.array([[81, -45, 45], [-45, 50, -15], [45, -15, 38]])
print('Определитель и главные миноры:', end=' ')
for i in range(2, -1, -1):
    print(round(np.linalg.det(a[:i+1, :i+1]), 3), end='  ')
print()    
print('Собственные значения:', end=' ')
for i in range(3):
    print(round(np.linalg.eigvals(a)[i], 3), end='  ')
```

Все условия выполняются. Произведем разложение на $L$ и $L^T$:

$$l_{11}=\sqrt{a_{11}}=\sqrt{81}=9,$$
$$l_{21}=\frac{a_{21}}{l_{11}}=\frac{-45}{9}=-5,$$
$$l_{31}=\frac{a_{31}}{l_{11}}=\frac{45}{9}=5,$$

$$l_{22}=\sqrt{a_{22}-l_{21}^{2}}=\sqrt{50 - 25}=5,$$
$$l_{32}=\frac{1}{l_{22}}\left ( a_{32}-l_{21}l_{31} \right)=\frac{1}{5}\left (-15 + 5\cdot5\right )=2,$$

$$l_{33}=\sqrt{a_{33}-l_{31}^{2}-l_{32}^{2}}=\sqrt{38-25-4}=3.$$

Получили матрицы:

$$L = \begin{pmatrix}
9 & 0 & 0 \\ 
-5 & 5 & 0 \\ 
5 & 2 & 3
\end{pmatrix}, 
\; \; 
L^{T} = \begin{pmatrix}
9 & -5 & 5 \\ 
0 & 5 & 2 \\ 
0 & 0 & 3
\end{pmatrix}.$$

Решим систему $Ly=b:$

$$\begin{cases}
9y_{1}=531, \\
-5y_{1}+5y_{2}=-460, \\
5y_{1}+2y_{2}+3y_{3}=193.
\end{cases}$$

$$y_{1} = 59,$$

$$y_{2}=-33,$$

$$y_{3}=-12.$$

И решим систему $L^{T}x=y:$

$$\begin{cases}
9x_{1}-5x_{2}+5x_{3}=59, \\
5x_{2}+2x_{3}=-33, \\
3x_{3}=-12.
\end{cases}$$

$$x_{3}=-4,$$

$$x_{2}=-5,$$

$$x_{1}=6.$$
