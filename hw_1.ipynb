# Линейное пространство. Часть 1.

### Задание №1.
Исследовать на линейную зависимость:

$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

__Решение:__   
Так как мы можем представить $f_4(x)$ как линейную комбинацию из $f_1(x), f_2(x), f_3(x)$:

$$x-e^x=(x+1)-1-e^x$$
$$f_4(x)=f_3(x)-f_2(x)-f_1(x)$$

То можем утверждать, что данные функции линейно зависимы.
___

### Задание №2.
Исследовать на линейную зависимость:

$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$$

__Решение:__   
Найдя квадрат суммы в $f_4(x)$:

$$f_4(x)=(x+1)^2=x^2+2x+1,$$

в этом примере мы также можем представить $f_4(x)$ как линейную комбинацию из $f_1(x), f_2(x), f_3(x)$:

$$f_4(x)=f_3(x)+2\cdot f_2(x)+0.5\cdot f_1(x).$$

Следовательно данные функции линейно зависимы.
___

### Задание №3.
Найти координаты вектора $x = (2, 3, 5)\in \mathbb{R}^{3}$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.

__Решение:__

$x = (2, 3, 5) = 0.5\cdot(0, 0, 10) + (2, 0, 0) + 3\cdot(0, 1, 0) = 0.5b_1 + b_2 + 3b_3$   
$x_b = (0.5, 1, 3)$
___

### Задание №4.
Найти координаты вектора $3x^{2}-2x+2\in\mathbb{R}^{3}[x]$:

а) в базисе $1$, $x$, $x^{2}$;

б) в базисе $x^{2}$, $x-1$, $1$.

__Решение:__

a) $3x^2-2x+2 = 2\cdot(1) - 2\cdot(x) + 3\cdot(x^2) = (2, -2, 3)$

б) $3x^2-2x+2 = 3\cdot(x^2) - 2\cdot(x-1) + 0\cdot(1) = (3, -2, 0)$
___

### Задание №5.
Установить, является ли линейным подпространством:

а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;
    
б) все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, \ldots, u_{n}\}$.

__Решение:__

__а)__ Возьмем два вектора: $a = (0, a_2, a_3), b = (b_1, 0, b_3)$, где $a_2, b_1$ - не равны 0.   
При их сложении: $a + b = (b_1, a_2, a_3 + b_3)$, получаем вектор, у которого ни одна из первых координат не равна 0. Следовательно такие векторы трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю, не являются линейным подпространством.

__б)__ И сумма векторов, и умноженный на число вектор, которые являются линейными комбинациями векторов $\{u_{1}, u_{2}, \ldots, u_{n}\}$, равны векторам, которые также представляют из себя линейные комбинации векторов $\{u_{1}, u_{2}, \ldots, u_{n}\}$. Поэтому заданные векторы являются линейным подпространством.

# Линейное пространство. Часть №2.

### Задание №1.
Найти скалярное произведение векторов $x, y \in \mathbb{R}$:   
а) $x=(0,-3, 6),~y=(-4, 7, 9);$   
б) $x=(7, -4, 0, 1),~y=(-3, 1, 11, 2).$

__Решение:__

__а)__ $(x, y) = 0\cdot(-4) + (-3)\cdot7 + 6\cdot9 = 33$   
__б)__ $(x, y) = 7\cdot(-3) + (-4)\cdot1 + 0\cdot11 + 1\cdot2 = -23$
___

### Задание №2.
Найти нормы векторов $a = (4, 2, 4)$ и $b = (12, 3, 4)$ и угол между ними.

__Решение:__

$\|a\|_1 = 4 + 2 + 4 = 10$   
$\|a\|_2 = \sqrt{4^2 + 2^2 + 4^2} = 6$   
$\|b\|_1 = 12 + 3 + 4 = 19$   
$\|b\|_2 = \sqrt{12^2 + 3^2 + 4^2} = 13$

$\displaystyle\cos\alpha = \frac{(a,b)}{\|a\|_2\|b\|_2} = \frac{4\cdot12 + 2\cdot3 + 4\cdot4}{6\cdot13} = \frac{35}{39}$
___

### Задание №3.
Будет ли линейное пространство евклидовым, если за скалярное произведение принять:   
а) произведение длин векторов;   
б) утроенное обычное скалярное произведение векторов?

__Решение:__

__а)__ Если заданы три __линейнонезависимых__ вектора $x, y, z$, то $|x+y|\cdot|z| \neq |x|\cdot|z|+|y|\cdot|z|$. Потому как длина суммарного вектора $x+y$ и сумма длин векторов $x$ и $y$ не равны. Третья аксиома не выполняется, следовательно произведение длин векторов за скалярное произведение мы принять не можем.

__б)__ Умножение обычного скалярного произведения векторов на любое число ни одну из 4-х аксиом не нарушит. Следовательно мы можем принять и утроенное скалярное произведение векторов за обычное.
___

### Задание №4.
Какие из нижеперечисленных векторов $a, b, c$ образуют ортонормированный базис в линейном пространстве $\mathbb{R}^{3}$:<br>
а) $(1,0,0),(0,0,1);$ <br>
б) $(1/\sqrt{2},-1/\sqrt{2},0),(1/\sqrt{2},1/\sqrt{2},0), (0,0,1);$<br>
в) $(1/2, -1/2, 0), (0, 1/2, 1/2), (0,0,1);$<br>
г) $(1,0,0),(0,1,0),(0,0,1)?$

Условия образования ортонормированного базиса:
- попарные скалярные произведения векторов равны 0,
- длины векторов равны 1.

__Решение:__   
__а)__ Два вектора не могут образовывать базис в трехмерном пространстве.

__б)__ Условия выполняются, это ортонормированный базис.   
$(a, b) = 1/\sqrt{2}\cdot1/\sqrt{2} - 1/\sqrt{2}\cdot1/\sqrt{2} + 0\cdot0 = 0$   
$(a, c) = 1/\sqrt{2}\cdot0 - 1/\sqrt{2}\cdot0 + 0\cdot1 = 0$   
$(b, c) = 1/\sqrt{2}\cdot0 + 1/\sqrt{2}\cdot0 + 0\cdot1 = 0$   
$|a| = \sqrt{(1/\sqrt{2})^2 + (-1/\sqrt{2})^2 + 0^2} = 1$   
$|b| = \sqrt{(1/\sqrt{2})^2 + (1/\sqrt{2})^2 + 0^2} = 1$   
$|c| = \sqrt{0^2 + 0^2 + 1^2} = 1$   

__в)__ Условие не выполняется, базис не ортонормированный.   
$(a, b) = 1/2\cdot0 - 1/2\cdot1/2 + 0\cdot1/2 = -1/4$   

__г)__ Ортонормированный базис из единичных векторов $i, j, k$.
